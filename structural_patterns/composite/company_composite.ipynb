```python
class Employee:
    def __init__(self, name, position):
        self.name = name
        self.position = position
        self.subordinates = []

    def add(self, employee):
        self.subordinates.append(employee)
    
    def show(self, indent=0):
        print(" " * indent + f"{self.position}: {self.name}")
        for s in self.subordinates:
            s.show(indent + 4)

        
ceo = Employee("Max", "CEO")
head_dev = Employee("Bob", "Head of Development")
dev1 = Employee("Charlie", "Developer")
dev2 = Employee("Anna", "Developer")
```


| Role          | Description                                                   |
| ------------- | ------------------------------------------------------------- |
| **Component** | Common interface for all objects in the hierarchy             |
| **Leaf**      | Represents individual objects (no children)                   |
| **Composite** | Represents containers that can hold children                  |
| **Client**    | Works with both leaf and composite through the same interface |
