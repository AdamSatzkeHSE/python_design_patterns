# SOLID Principles
Acronym by Robert C. Martin
Represents *five principles* for writing good object-oriented software.

**S: Single Responsability**

"A class should have only one reason to change"

**O: Open/Closed**

"Open for extension, closed for modification"

**L: Liskov Substitution**

"Subclasses should be usable anywhere where their base class is"

**I: Interface Segregation**

"Don't force classes to implement what they don't need"

**D: Dependency Inversion**

"Depend on abstractions, not concretions"
