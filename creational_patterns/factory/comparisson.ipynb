| Aspect         | Factory Method                       | Abstract Factory                          |
| -------------- | ------------------------------------ | ----------------------------------------- |

| Implementation | Subclasses override a factory method | Factories group creation methods          |
| Example        | Truck vs Ship                        | Victorian Chair + Victorian Table         |
